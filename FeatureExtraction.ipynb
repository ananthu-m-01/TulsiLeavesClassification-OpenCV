```python
import cv2
import os
import matplotlib.pyplot as plt
import pandas as pd

# Define dataset paths
good_folder = "dataset/Good"
bad_folder = "dataset/Bad"

# Function to load one sample image and count images
def load_sample_and_count(folder):
    images = os.listdir(folder)
    total_count = len(images)
    
    sample_img = None
    if images:
        sample_img_path = os.path.join(folder, images[0])
        sample_img = cv2.imread(sample_img_path)
        if sample_img is not None:
            sample_img = cv2.cvtColor(sample_img, cv2.COLOR_BGR2RGB)
    
    return total_count, sample_img

# Get counts and sample images
good_count, good_sample = load_sample_and_count(good_folder)
bad_count, bad_sample = load_sample_and_count(bad_folder)

# Display sample images
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(good_sample)
plt.axis("off")
plt.title("Good Leaf Sample")

plt.subplot(1, 2, 2)
plt.imshow(bad_sample)
plt.axis("off")
plt.title("Bad Leaf Sample")
plt.show()

# Show count
df = pd.DataFrame({
    "Category": ["Good Leaves", "Bad Leaves"],
    "Total Count": [good_count, bad_count]
})
print(df)
```

```python
import cv2
import os
import matplotlib.pyplot as plt
import random

def display_good_bad_edges(good_folder, bad_folder):
    # Function to load a random image and apply edge detection
    def load_random_edge(folder):
        images = os.listdir(folder)
        random_image = random.choice(images)
        image_path = os.path.join(folder, random_image)

        # Read image
        image = cv2.imread(image_path)

        # Convert to grayscale
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Apply Canny edge detection
        edges = cv2.Canny(gray, threshold1=100, threshold2=200)

        return edges

    # Get edges for good and bad samples
    good_edges = load_random_edge(good_folder)
    bad_edges = load_random_edge(bad_folder)

    # Display edge images
    plt.figure(figsize=(8, 4))

    plt.subplot(1, 2, 1)
    plt.imshow(good_edges, cmap='gray')
    plt.axis("off")
    plt.title("Good Leaf Edges")

    plt.subplot(1, 2, 2)
    plt.imshow(bad_edges, cmap='gray')
    plt.axis("off")
    plt.title("Bad Leaf Edges")

    plt.show()
```

```python
display_good_bad_edges("dataset/Good", "dataset/Bad")
```

```python
import cv2
import os
import matplotlib.pyplot as plt
import random

def display_good_bad_grayscale(good_folder, bad_folder):
    # Function to load random image and convert it to grayscale
    def load_random_and_display(folder):
        images = os.listdir(folder)
        random_image = random.choice(images)
        image_path = os.path.join(folder, random_image)
        
        # Read image
        image = cv2.imread(image_path)
        
        # Convert to grayscale
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
        return gray_image
    
    # Load random good and bad leaf images
    good_gray = load_random_and_display(good_folder)
    bad_gray = load_random_and_display(bad_folder)
    
    # Display grayscale images
    plt.figure(figsize=(8, 4))

    plt.subplot(1, 2, 1)
    plt.imshow(good_gray, cmap='gray')
    plt.axis("off")
    plt.title("Good Leaf Sample (Grayscale)")

    plt.subplot(1, 2, 2)
    plt.imshow(bad_gray, cmap='gray')
    plt.axis("off")
    plt.title("Bad Leaf Sample (Grayscale)")

    plt.show()
```

```python
import cv2
import os
import matplotlib.pyplot as plt
import random

def display_good_bad_contrast(good_folder, bad_folder):
    # Function to load a random image and apply contrast enhancement
    def load_random_contrast(folder):
        images = os.listdir(folder)
        random_image = random.choice(images)
        image_path = os.path.join(folder, random_image)

        # Read image and convert to grayscale
        image = cv2.imread(image_path)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Apply histogram equalization
        contrast_enhanced = cv2.equalizeHist(gray)

        return contrast_enhanced

    # Load and enhance contrast
    good_contrast = load_random_contrast(good_folder)
    bad_contrast = load_random_contrast(bad_folder)

    # Display contrast-enhanced images
    plt.figure(figsize=(8, 4))

    plt.subplot(1, 2, 1)
    plt.imshow(good_contrast, cmap='gray')
    plt.axis("off")
    plt.title("Good Leaf (Contrast Enhanced)")

    plt.subplot(1, 2, 2)
    plt.imshow(bad_contrast, cmap='gray')
    plt.axis("off")
    plt.title("Bad Leaf (Contrast Enhanced)")

    plt.show()
```

```python
display_good_bad_contrast("dataset/Good", "dataset/Bad")
```

```python
from skimage.feature import local_binary_pattern

def display_good_bad_texture_lbp(good_folder, bad_folder):
    def load_random_lbp(folder):
        images = os.listdir(folder)
        image_path = os.path.join(folder, random.choice(images))
        image = cv2.imread(image_path)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Apply LBP
        lbp = local_binary_pattern(gray, P=8, R=1, method='uniform')
        return lbp

    good_lbp = load_random_lbp(good_folder)
    bad_lbp = load_random_lbp(bad_folder)

    plt.figure(figsize=(8, 4))

    plt.subplot(1, 2, 1)
    plt.imshow(good_lbp, cmap='gray')
    plt.axis("off")
    plt.title("Good Leaf (LBP Texture)")

    plt.subplot(1, 2, 2)
    plt.imshow(bad_lbp, cmap='gray')
    plt.axis("off")
    plt.title("Bad Leaf (LBP Texture)")

    plt.show()
```

```python
display_good_bad_texture_lbp("dataset/Good", "dataset/Bad")
```

```python
def display_good_bad_histogram(good_folder, bad_folder):
    def load_random_hist(folder):
        images = os.listdir(folder)
        image_path = os.path.join(folder, random.choice(images))
        image = cv2.imread(image_path)
        hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
        hist = cv2.calcHist([hsv], [0], None, [180], [0, 180])
        hist = cv2.normalize(hist, hist).flatten()
        return hist

    good_hist = load_random_hist(good_folder)
    bad_hist = load_random_hist(bad_folder)

    plt.figure(figsize=(10, 4))
    plt.plot(good_hist, label='Good Leaf', color='green')
    plt.plot(bad_hist, label='Bad Leaf', color='red')
    plt.title("Hue Histogram Comparison")
    plt.xlabel("Hue Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
display_good_bad_histogram("dataset/Good", "dataset/Bad")
```

```python
import cv2
import os
import random
import matplotlib.pyplot as plt
from skimage.feature import local_binary_pattern

def display_all_features_comparison(good_folder, bad_folder):
    def load_random_image(folder):
        image_path = os.path.join(folder, random.choice(os.listdir(folder)))
        image = cv2.imread(image_path)
        return image

    def convert_gray(image):
        return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    def get_edges(gray_img):
        return cv2.Canny(gray_img, 100, 200)

    def get_contrast(gray_img):
        return cv2.equalizeHist(gray_img)

    def get_lbp(gray_img):
        return local_binary_pattern(gray_img, P=8, R=1, method='uniform')

    def get_hue_histogram(image):
        hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
        hist = cv2.calcHist([hsv], [0], None, [180], [0, 180])
        hist = cv2.normalize(hist, hist).flatten()
        return hist

    # Load images
    good_img = load_random_image(good_folder)
    bad_img = load_random_image(bad_folder)

    # Convert
    good_gray = convert_gray(good_img)
    bad_gray = convert_gray(bad_img)

    # Processed Features
    good_edge = get_edges(good_gray)
    bad_edge = get_edges(bad_gray)

    good_contrast = get_contrast(good_gray)
    bad_contrast = get_contrast(bad_gray)

    good_lbp = get_lbp(good_gray)
    bad_lbp = get_lbp(bad_gray)

    good_hist = get_hue_histogram(good_img)
    bad_hist = get_hue_histogram(bad_img)

    # Plotting
    fig, axs = plt.subplots(5, 2, figsize=(12, 14))

    # 1. Original
    axs[0, 0].imshow(cv2.cvtColor(good_img, cv2.COLOR_BGR2RGB))
    axs[0, 0].set_title("Good Leaf - Color")
    axs[0, 1].imshow(cv2.cvtColor(bad_img, cv2.COLOR_BGR2RGB))
    axs[0, 1].set_title("Bad Leaf - Color")

    # 2. Grayscale
    axs[1, 0].imshow(good_gray, cmap='gray')
    axs[1, 0].set_title("Good Leaf - Grayscale")
    axs[1, 1].imshow(bad_gray, cmap='gray')
    axs[1, 1].set_title("Bad Leaf - Grayscale")

    # 3. Edges
    axs[2, 0].imshow(good_edge, cmap='gray')
    axs[2, 0].set_title("Good Leaf - Edges")
    axs[2, 1].imshow(bad_edge, cmap='gray')
    axs[2, 1].set_title("Bad Leaf - Edges")

    # 4. Contrast
    axs[3, 0].imshow(good_contrast, cmap='gray')
    axs[3, 0].set_title("Good Leaf - Contrast Enhanced")
    axs[3, 1].imshow(bad_contrast, cmap='gray')
    axs[3, 1].set_title("Bad Leaf - Contrast Enhanced")

    # 5. LBP Texture
    axs[4, 0].imshow(good_lbp, cmap='gray')
    axs[4, 0].set_title("Good Leaf - LBP Texture")
    axs[4, 1].imshow(bad_lbp, cmap='gray')
    axs[4, 1].set_title("Bad Leaf - LBP Texture")

    # Hide axes
    for ax in axs.ravel():
        ax.axis('off')

    plt.tight_layout()
    plt.show()

    # Separate Plot for Histogram
    plt.figure(figsize=(10, 4))
    plt.plot(good_hist, label='Good Leaf Hue', color='green')
    plt.plot(bad_hist, label='Bad Leaf Hue', color='red')
    plt.title("Hue Histogram Comparison")
    plt.xlabel("Hue Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
display_all_features_comparison("dataset/Good", "dataset/Bad")
```

```python
import cv2
import os
import random
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from skimage.feature import local_binary_pattern

def display_and_save_feature_report(good_folder, bad_folder, output_folder='reports'):
    os.makedirs(output_folder, exist_ok=True)

    def load_random_image(folder):
        image_path = os.path.join(folder, random.choice(os.listdir(folder)))
        image = cv2.imread(image_path)
        return image

    def convert_gray(image):
        return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    def get_edges(gray_img):
        return cv2.Canny(gray_img, 100, 200)

    def get_contrast(gray_img):
        return cv2.equalizeHist(gray_img)

    def get_lbp(gray_img):
        return local_binary_pattern(gray_img, P=8, R=1, method='uniform')

    def get_hue_histogram(image):
        hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
        hist = cv2.calcHist([hsv], [0], None, [180], [0, 180])
        hist = cv2.normalize(hist, hist).flatten()
        return hist

    # Load sample images
    good_img = load_random_image(good_folder)
    bad_img = load_random_image(bad_folder)

    good_gray = convert_gray(good_img)
    bad_gray = convert_gray(bad_img)

    good_edge = get_edges(good_gray)
    bad_edge = get_edges(bad_gray)

    good_contrast = get_contrast(good_gray)
    bad_contrast = get_contrast(bad_gray)

    good_lbp = get_lbp(good_gray)
    bad_lbp = get_lbp(bad_gray)

    good_hist = get_hue_histogram(good_img)
    bad_hist = get_hue_histogram(bad_img)

    feature_titles = [
        ("Original", good_img, bad_img, 'RGB'),
        ("Grayscale", good_gray, bad_gray, 'gray'),
        ("Edges", good_edge, bad_edge, 'gray'),
        ("Contrast", good_contrast, bad_contrast, 'gray'),
        ("LBP Texture", good_lbp, bad_lbp, 'gray')
    ]

    pdf_path = os.path.join(output_folder, "Tulsi_Leaf_Feature_Comparison_Report.pdf")
    with PdfPages(pdf_path) as pdf:
        for title, good_feat, bad_feat, cmap in feature_titles:
            fig, axs = plt.subplots(1, 2, figsize=(10, 4))
            axs[0].imshow(good_feat, cmap=None if cmap == 'RGB' else cmap)
            axs[0].set_title(f"Good - {title}")
            axs[0].axis('off')

            axs[1].imshow(bad_feat, cmap=None if cmap == 'RGB' else cmap)
            axs[1].set_title(f"Bad - {title}")
            axs[1].axis('off')

            plt.tight_layout()
            pdf.savefig(fig)
            plt.close()

        # Hue Histogram
        fig = plt.figure(figsize=(8, 4))
        plt.plot(good_hist, label="Good Leaf Hue", color='green')
        plt.plot(bad_hist, label="Bad Leaf Hue", color='red')
        plt.title("Hue Histogram Comparison")
        plt.xlabel("Hue Value")
        plt.ylabel("Frequency")
        plt.legend()
        plt.grid(True)
        pdf.savefig(fig)
        plt.close()

    print(f"[✔] Report saved at: {pdf_path}")
```

```python
display_and_save_feature_report("dataset/Good", "dataset/Bad")
```
